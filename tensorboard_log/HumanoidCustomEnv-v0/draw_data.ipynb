# 从tensorboard的数据绘制图像 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_name_1 = "download\\run-2e6_t1_cpu6_ppo_HumanoidCustomEnv-v0_1-tag-rollout_ep_rew_mean.csv"
file_name_2 = "download\\run-2e6_t2_cpu6_ppo_HumanoidCustomEnv-v0_1-tag-rollout_ep_rew_mean.csv"

f1 = pd.read_csv(file_name_1)
f2 = pd.read_csv(file_name_2)
```

```python
f1_steps = np.array(f1["Step"][:])
f2_steps = np.array(f2["Step"][:])
f1_rewards = np.array(f1["Value"][:])
f2_rewards = np.array(f2["Value"][:])
plt.plot(f1_steps, f1_rewards)
plt.plot(f2_steps, f2_rewards)
plt.title("PPO-HumanoidCustom")
plt.legend(["2e6","5e6"])
plt.show()
```

```python

sac2e6t4 = pd.read_csv("download\\run-2e6_t4_cpu5_sac_HumanoidCustomEnv-v0_1-tag-rollout_ep_rew_mean.csv")
sac2e6t4_steps = np.array(sac2e6t4["Step"][:])
sac2e6t4_values = np.array(sac2e6t4["Value"][:])
plt.plot(f1_steps, f1_rewards)
plt.plot(sac2e6t4_steps, sac2e6t4_values)
plt.title("Algo Compare")
plt.legend(["PPO","SAC"])
plt.show()
```

```python
sac2e6t8 = pd.read_csv("download\\run-2e6_t8_cpu6_sac_HumanoidCustomEnv-v0_1-tag-rollout_ep_rew_mean.csv")
sac2e6t8_steps = np.array(sac2e6t8["Step"][:])
sac2e6t8_values = np.array(sac2e6t8["Value"][:])
plt.plot(sac2e6t4_steps, sac2e6t4_values) # sac 2e6次迭代
plt.plot(sac2e6t8_steps, sac2e6t8_values) # sac 5e6次迭代
plt.title("SAC")
plt.legend(["2E6","5E6"])
plt.show()
```
